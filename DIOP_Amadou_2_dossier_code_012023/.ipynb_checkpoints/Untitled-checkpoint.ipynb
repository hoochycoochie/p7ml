```python
import pandas as pd
import numpy as np
from functions import functions as utils2
```

```python
path='data/'
```

```python
df_test =utils2.reduce_memory_usage( pd.read_csv(path+'application_test.csv'))
 
df_bureau =utils2.reduce_memory_usage( pd.read_csv(path+'bureau.csv'))
df_pos_cash_balance = utils2.reduce_memory_usage(pd.read_csv(path+'POS_CASH_balance.csv'))
df_credit_card_balance=utils2.reduce_memory_usage(pd.read_csv(path+'credit_card_balance.csv'))

df_previous_application = utils2.reduce_memory_usage(pd.read_csv(path+'previous_application.csv'))
df_installments_payments = utils2.reduce_memory_usage(pd.read_csv(path+'installments_payments.csv'))
```

```python
df_test
```

```python
def client_file(SK_ID_CURR):
    df_test1=df_test[df_test['SK_ID_CURR']==SK_ID_CURR]
    df_test1.to_csv('test/1/df_main.csv',index=False)


    df_bureau1=df_bureau[df_bureau['SK_ID_CURR']==SK_ID_CURR]
    df_bureau1.to_csv('test/1/df_bureau.csv',index=False)



    df_pos_cash_balance1=df_pos_cash_balance[df_pos_cash_balance['SK_ID_CURR']==SK_ID_CURR]
    df_pos_cash_balance1.to_csv('test/1/df_pos_cash_balance.csv',index=False)


    df_credit_card_balance1=df_credit_card_balance[df_credit_card_balance['SK_ID_CURR']==SK_ID_CURR]
    df_credit_card_balance1.to_csv('test/1/df_credit_card_balance.csv',index=False)



    df_previous_application1=df_previous_application[df_previous_application['SK_ID_CURR']==SK_ID_CURR]
    df_previous_application1.to_csv('test/1/df_previous_application.csv',index=False)


    df_installments_payments1=df_installments_payments[df_installments_payments['SK_ID_CURR']==SK_ID_CURR]
    df_installments_payments1.to_csv('test/1/df_installments_payments.csv',index=False)
```

```python
client_file(100001)
```
