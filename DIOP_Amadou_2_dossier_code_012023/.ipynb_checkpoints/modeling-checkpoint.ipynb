```python

def evaluate_model(model,X_test,y_test):
    labels = model.predict(X_test)

    y_pred_proba = model.predict_proba(X_test)[::, 1]
    fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
    auc = roc_auc_score(y_test, y_pred_proba)

    # create ROC curve
    plt.plot(fpr, tpr, label="AUC=" + str(auc))
    plt.ylabel("True Positive Rate")
    plt.xlabel("False Positive Rate")
    plt.legend(loc=4)
    plt.show()
    y_true = y_test
    y_pred = model.predict(X_test)

    # plot_cm(y_true, y_pred,labels_inverse)

    # skplt.metrics.plot_confusion_matrix(
    # y_true_label,
    # y_pred_label,
    # figsize=(12,12))

    scoring_func.plot_cm(y_true, y_pred)
    
def find_doc(cols):
    return(doc[doc['Row'].isin(cols)][['Row','Description']])
def correlation_with_target_top_n(df,target):
    '''
    Calculate the correltation
    '''
    df_corr= df.corr()[target].sort_values(ascending=False)
    #print(n,'Positive Correlations:\n', df_corr.tail(n))
    #print('\n',n,' Negative Correlations:\n', df_corr.head(n))
    
    return df_corr

def one_hot_encoding_dataframe(df):
    '''
    one hot encoding 
    '''
    original_columns = list(df.columns)
    cat_columns=[x for x in df.columns if df[x].dtype == 'object']
    df=pd.get_dummies(df,columns=cat_columns,dummy_na= False)
    new_added_columns=list(set(df.columns).difference(set(original_columns)))
    return df,new_added_columns,df.columns


def pca_features_importance(pca,cols):
    # number of components
    n_pcs= pca.components_.shape[0]

    # get the index of the most important feature on EACH component
    # LIST COMPREHENSION HERE
    most_important = [np.abs(pca.components_[i]).argmax() for i in range(n_pcs)]

    initial_feature_names = cols
    # get the names
    most_important_names = [initial_feature_names[most_important[i]] for i in range(n_pcs)]

    # LIST COMPREHENSION HERE AGAIN
    dic = {'PC{}'.format(i): most_important_names[i] for i in range(n_pcs)}

    # build the dataframe
    df_pca_features = pd.DataFrame(dic.items())
    return df_pca_features
    
```

## 2 <u>Clustering kmeans</u>

```python

# take too much long to compute from k=(2,10) , best k=5

kmeans_visualizer = KElbowVisualizer(KMeans(random_state=random_state),K=(2,9))
kmeans_visualizer.fit(df_pca.values)
kmeans_visualizer.show()
 
#from IPython.display import Image
#Image(filename='kmeans.png') 
```

* selon la méthode du coude, k=5 est le nombre de classes optimal

```python

K=kmeans_visualizer.elbow_value_
kmeans = KMeans(n_clusters=K,random_state=random_state)
kmeans.fit(df_pca.values)
## save kmeans instance for next test sample
 

joblib.dump(kmeans, fastapi_path+'kmeans.bin',compress=True)
joblib.dump(kmeans, streamlit_path+'kmeans.bin',compress=True)
```

## 3 <u>Smote équilibrage des classes</u>

* Etant donné le déséquilibre de classe observé sur la variable target, nous recourrons à l'algorithme smote (Synthetic minority oversampling technique) pour palier ce déséquilibre sur l'échantillon d'entrainement

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=random_state,stratify=y)
```

```python
 
X_res, y_res =scoring_func.smotify(X_train,y_train,k_neighbors=3,sampling_strategy='minority',random_state=random_state)
```

* resampling de la classe minoritaire à hauteur de la classe majoritaire (211989)
